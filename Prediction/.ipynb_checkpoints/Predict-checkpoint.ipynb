```python
def preproc(df):
    #### ---------CLICK_POSISIONS-----------####
    df['global_url_ctr'] = df['global_url_clicks'] / df['global_url_shows']
    df['global_url_pos1ctr'] = df['global_url_pos1clicks'] / df['global_url_pos1shows']
    df['global_url_pos2ctr'] = df['global_url_pos2clicks'] / df['global_url_pos2shows']
    df['global_url_pos3ctr'] = df['global_url_pos3clicks'] / df['global_url_pos3shows']
    df['global_url_pos4ctr'] = df['global_url_pos4clicks'] / df['global_url_pos4shows']
    df['global_url_pos5ctr'] = df['global_url_pos5clicks'] / df['global_url_pos5shows']
    df['global_url_pos6ctr'] = df['global_url_pos6clicks'] / df['global_url_pos6shows']
    df['global_url_pos7ctr'] = df['global_url_pos7clicks'] / df['global_url_pos7shows']
    df['global_url_pos8ctr'] = df['global_url_pos8clicks'] / df['global_url_pos8shows']
    df['global_url_pos9ctr'] = df['global_url_pos9clicks'] / df['global_url_pos9shows']   
    df['global_url_pos10ctr'] = df['global_url_pos10clicks'] / df['global_url_pos10shows']
    
    df['query_url_ctr'] = df['query_url_clicks'] / df['query_url_shows']
    df['query_url_pos1ctr'] = df['query_url_pos1clicks'] / df['query_url_pos1shows']
    df['query_url_pos2ctr'] = df['query_url_pos2clicks'] / df['query_url_pos2shows']
    df['query_url_pos3ctr'] = df['query_url_pos3clicks'] / df['query_url_pos3shows']
    df['query_url_pos4ctr'] = df['query_url_pos4clicks'] / df['query_url_pos4shows']
    df['query_url_pos5ctr'] = df['query_url_pos5clicks'] / df['query_url_pos5shows']
    df['query_url_pos6ctr'] = df['query_url_pos6clicks'] / df['query_url_pos6shows']
    df['query_url_pos7ctr'] = df['query_url_pos7clicks'] / df['query_url_pos7shows']
    df['query_url_pos8ctr'] = df['query_url_pos8clicks'] / df['query_url_pos8shows']
    df['query_url_pos9ctr'] = df['query_url_pos9clicks'] / df['query_url_pos9shows']
    df['query_url_pos10ctr'] = df['query_url_pos10clicks'] / df['query_url_pos10shows']
    
    df['global_host_ctr'] = df['global_host_clicks'] / df['global_host_shows']
    df['global_host_pos1ctr'] = df['global_host_pos1clicks'] / df['global_host_pos1shows']
    df['global_host_pos2ctr'] = df['global_host_pos2clicks'] / df['global_host_pos2shows']
    df['global_host_pos3ctr'] = df['global_host_pos3clicks'] / df['global_host_pos3shows']
    df['global_host_pos4ctr'] = df['global_host_pos4clicks'] / df['global_host_pos4shows']
    df['global_host_pos5ctr'] = df['global_host_pos5clicks'] / df['global_host_pos5shows']
    df['global_host_pos6ctr'] = df['global_host_pos6clicks'] / df['global_host_pos6shows']
    df['global_host_pos7ctr'] = df['global_host_pos7clicks'] / df['global_host_pos7shows']
    df['global_host_pos8ctr'] = df['global_host_pos8clicks'] / df['global_host_pos8shows']
    df['global_host_pos9ctr'] = df['global_host_pos9clicks'] / df['global_host_pos9shows']   
    df['global_host_pos10ctr'] = df['global_host_pos10clicks'] / df['global_host_pos10shows']
    
    df['query_host_ctr'] = df['query_host_clicks'] / df['query_host_shows']
    df['query_host_pos1ctr'] = df['query_host_pos1clicks'] / df['query_host_pos1shows']
    df['query_host_pos2ctr'] = df['query_host_pos2clicks'] / df['query_host_pos2shows']
    df['query_host_pos3ctr'] = df['query_host_pos3clicks'] / df['query_host_pos3shows']
    df['query_host_pos4ctr'] = df['query_host_pos4clicks'] / df['query_host_pos4shows']
    df['query_host_pos5ctr'] = df['query_host_pos5clicks'] / df['query_host_pos5shows']
    df['query_host_pos6ctr'] = df['query_host_pos6clicks'] / df['query_host_pos6shows']
    df['query_host_pos7ctr'] = df['query_host_pos7clicks'] / df['query_host_pos7shows']
    df['query_host_pos8ctr'] = df['query_host_pos8clicks'] / df['query_host_pos8shows']
    df['query_host_pos9ctr'] = df['query_host_pos9clicks'] / df['query_host_pos9shows']
    df['query_host_pos10ctr'] = df['query_host_pos10clicks'] / df['query_host_pos10shows']
    
    #### ---------TIMERS-----------####
    df['global_avg_time'] = df['global_time'] / df['global_url_clicks']
    df['query_url_avg_time'] = df['query_url_time'] / df['query_url_shows']
    df['global_host_avg_time'] = df['global_host_time'] / df['global_host_clicks']
    df['query_host_avg_time'] = df['query_host_time'] / df['query_host_shows']
    
    #### ---------RELATIVE_CLICKS-----------####
    df['global_click_above'] = df['global_above_clicks'] / df['global_url_shows']
    df['global_click_below'] = df['global_below_clicks'] / df['global_url_shows']
    df['global_click_prev'] = df['global_prev_clicks'] / df['global_url_shows']
    df['global_click_next'] = df['global_next_clicks'] / df['global_url_shows']
    
    df['query_url_click_above'] = df['query_url_above_clicks'] / df['query_url_shows']
    df['query_url_click_below'] = df['query_url_below_clicks'] / df['query_url_shows']
    df['query_url_click_prev'] = df['query_url_prev_clicks'] / df['query_url_shows']
    df['query_url_click_next'] = df['query_url_next_clicks'] / df['query_url_shows']
    
    df['global_host_click_above'] = df['global_host_above_clicks'] / df['global_host_shows']
    df['global_host_click_below'] = df['global_host_below_clicks'] / df['global_host_shows']
    df['global_host_click_prev'] = df['global_host_prev_clicks'] / df['global_host_shows']
    df['global_host_click_next'] = df['global_host_next_clicks'] / df['global_host_shows']
    
    df['query_host_click_above'] = df['query_host_above_clicks'] / df['query_host_shows']
    df['query_host_click_below'] = df['query_host_below_clicks'] / df['query_host_shows']
    df['query_host_click_prev'] = df['query_host_prev_clicks'] / df['query_host_shows']
    df['query_host_click_next'] = df['query_host_next_clicks'] / df['query_host_shows']
    
    #### ---------CLICK_MODELS-----------####
    df['global_url_pos_based'] = df['global_url_clicks'] / df['global_url_pos_based_shows']
    df['global_url_cascade'] = df['global_url_cascade_clicks'] / df['global_url_cascade_shows']
    df['global_url_dbn_a'] = (df['global_url_dbn_clicks'] + 0.1) / (df['global_url_dbn_shows'] + 2 * 0.1)
    df['global_url_dbn_s'] = (df['global_url_dbn_satisfied'] + 0.1) / (df['global_url_dbn_clicks'] + 2 * 0.1)
    df['global_url_dbn'] = df['global_url_dbn_a'] * df['global_url_dbn_s']
    
    df['query_url_pos_based'] = df['query_url_clicks'] / df['query_url_pos_based_shows']
    df['query_url_cascade'] = df['query_url_cascade_clicks'] / df['query_url_cascade_shows']
    df['query_url_dbn_a'] = (df['query_url_dbn_clicks'] + 0.1) / (df['query_url_dbn_shows'] + 2 * 0.1)
    df['query_url_dbn_s'] = (df['query_url_dbn_satisfied'] + 0.1) / (df['query_url_dbn_clicks'] + 2 * 0.1)
    df['query_url_dbn'] = df['query_url_dbn_a'] * df['query_url_dbn_s']

    df['global_host_pos_based'] = df['global_host_clicks'] / df['global_host_pos_based_shows']
    df['global_host_cascade'] = df['global_host_cascade_clicks'] / df['global_host_cascade_shows']
    df['global_host_dbn_a'] = (df['global_host_dbn_clicks'] + 0.1) / (df['global_host_dbn_shows'] + 2 * 0.1)
    df['global_host_dbn_s'] = (df['global_host_dbn_satisfied'] + 0.1) / (df['global_host_dbn_clicks'] + 2 * 0.1)
    df['global_host_dbn'] = df['global_host_dbn_a'] * df['global_host_dbn_s']

    df['query_host_pos_based'] = df['query_host_clicks'] / df['query_host_pos_based_shows']
    df['query_host_cascade'] = df['query_host_cascade_clicks'] / df['query_host_cascade_shows']
    df['query_host_dbn_a'] = (df['query_host_dbn_clicks'] + 0.1) / (df['query_host_dbn_shows'] + 2 * 0.1)
    df['query_host_dbn_s'] = (df['query_host_dbn_satisfied'] + 0.1) / (df['query_host_dbn_clicks'] + 2 * 0.1)
    df['query_host_dbn'] = df['query_host_dbn_a'] * df['query_host_dbn_s']
    
    return df
```

```python
from catboost import *
features = list(train_df.columns[4:])
pool = Pool(
    data=train_df[features].values,
    label=train_df.mark.values / 5,
    group_id=train_df.query_id.values
)
```

```python
model_params = {
    'iterations': 7000,
    'custom_metric': ['NDCG:top=5'],
    'loss_function': 'YetiRankPairwise',
    'verbose': True,
    'random_seed': 1337,
    'task_type': 'GPU'
}
```

```python
model = CatBoost(model_params)
model.fit(pool, eval_set=[pool])
```

```python
test_pool = Pool(data=test_df[features].values)
out = pd.DataFrame()
out["query_id"] = test_df["query_id"]
out["pred_mark"] = model.predict(test_pool)
out["url_id"] = test_df["url_id"]
result = out.sort_values(by=['query_id',"pred_mark"], ascending = [True, False])
```

```python
with open('sub.txt', 'w') as f:
    f.write("QueryId,DocumentId\n")
    for query_id in set(result.query_id.tolist()):
        cur_urls = result[result.query_id==query_id].url_id.tolist()
        counter = 0
        used_urls = set()
        for url_id in cur_urls:
            counter += 1
            used_urls.add(url_id)
            f.write("{},{}\n".format(query_id,url_id))
            if counter == 5: 
                break
```

```python
weight = model.get_feature_importance(pool, verbose=True)
```

```python
top_features = []
top_weights = []
for f in np.argsort(weight)[::-1]:
    top_features.append(features[f])
    top_weights.append(weight[f])
```
